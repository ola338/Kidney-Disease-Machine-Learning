```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import seaborn as sns 
```

```python
dataset = pd.read_csv("new_model.csv", header=0)  
print(dataset.head)  
X = dataset.values[:,0:13]   # Zmienne objaśniające
y = dataset['Class']     # zmienna grupująca/zależna, 0 - brak ryzyka choroby, 1 - duże ryzyko
print(dataset.columns[:-1])
```

```python
# autoskalowanie zmiennych
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X) 
```

```python
# PCA i transformacja X do głównych składowych (PC)
pca = PCA(n_components=13)
pca.fit(X) 
PC = pca.transform(X)
```

```python

print('Procent opisywanej zmienności /%: ', 100 * pca.explained_variance_ratio_)   
sns.barplot(x=np.arange(1,14), y= 100 * pca.explained_variance_ratio_, color='r')  # wykres słupkowy wszystkich składowych
plt.xlabel('Numer głównej składowej (PC)')   
plt.ylabel('Opisywana zmienność / %')    
plt.show()   
```

```python
# Rzut obiektów na płaszczyznę PC1/PC2
print('\n Wyniki:')
sns.scatterplot(x=PC[:,0], y=PC[:,1], palette='dark', hue=y, legend='full', s=80)
plt.xlabel('PC1 / '+'% 4.2f' % (100*pca.explained_variance_ratio_)[0] + ' %')
plt.ylabel('PC2 / '+'% 4.2f' % (100*pca.explained_variance_ratio_)[1] + ' %')
plt.legend()
plt.show() 
```

```python
pca = PCA(n_components=6)   # zamiana na 6 składowych - istotne główne składowe
pca.fit(X) 
PC = pca.transform(X)
```

```python
print('Wagi 6 głównych składowych: ') 
print(pca.components_[0:6,:] )
df_comp = pd.DataFrame(pca.components_, columns = dataset.columns[:-1])  # mapa cieplna obrazująca ładunki składowych
plt.figure(figsize =(14, 6))
sns.heatmap(df_comp)   
plt.show()
```
